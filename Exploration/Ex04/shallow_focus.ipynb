## 사람, 동물의 segmentation color 확인

### 컬러맵 만들기

```python
#컬러맵 만들기 
colormap = np.zeros((256, 3), dtype = int)
ind = np.arange(256, dtype=int)

for shift in reversed(range(8)):
    for channel in range(3):
        colormap[:, channel] |= ((ind >> channel) & 1) << shift
    ind >>= 3

colormap[:20] #생성한 20개의 컬러맵 출력
```

### 사람과 동물 색상 정리

```python
human_color = (128,128,192)
animal_colors = [(0,0,64), (0,128,64), (128,0,64), (128,0,192), (0,64,128)]
```

## 모델 로드

기존에는 모델 로드를 인물모드 함수 내에서 구현했는데, 관련 경고가 발생하여 모델 로드는 함수 밖에서 1번만 실행하는 것이 효율적이라는 것을 배웠습니다!

```python
# 모델 위치
model_dir = os.getenv('HOME')+'/aiffel/AIFFEL_quest_rs/Exploration/Ex04/models' 
model_file = os.path.join(model_dir, 'deeplabv3_xception_tf_dim_ordering_tf_kernels.h5') 

# 모델 로드
model = semantic_segmentation()
model.load_pascalvoc_model(model_file)
```

## 인물모드 사진 제작 함수 구현

```python
# 원본 이미지, 배경 이미지, 라벨링 이름을 넣으면 인물모드 사진 생성
# save_path에 위치 넣으면 합성된 png 파일 저장
def changeBackground(img_path, bg_path=None, label_name='person', save_path=None):
    img_orig = cv2.imread(img_path)
    
    # 배경 설정
    if bg_path is None:
        bg = cv2.blur(img_orig, (13,13))
    else:
        bg_orig = cv2.imread(bg_path)
        bg = cv2.resize(bg_orig, (img_orig.shape[1], img_orig.shape[0]))
        
    # 모델 사용
    segvalues, output = model.segmentAsPascalvoc(img_path)
        
    # 라벨 이름에 따른 세그먼트 맵 생성
    if label_name == 'person':
        seg_map = np.all(output==human_color, axis=-1)
    elif label_name == 'animal':
        seg_map = np.any(
            np.stack([np.all(output == color, axis=-1) for color in animal_colors]),
            axis=0
        )
    else:
        raise ValueError("label_name은 'person' 또는 'animal'만 가능합니다.")
    
    # 마스크 처리
    img_mask = seg_map.astype(np.uint8) * 255
    img_mask_color = cv2.cvtColor(img_mask, cv2.COLOR_GRAY2BGR)

    # 배경과 합성
    img_concat = np.where(img_mask_color == 255, img_orig, bg)
    
    # 저장
    if save_path:
        cv2.imwrite(save_path, img_concat)
    
    # 결과 시각화
    plt.imshow(cv2.cvtColor(img_concat, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
    
    return
```

## 인물모드 사진 도출

### 인물 배경 블러

```python
# 인물 이미지 위치 불러오기
img_path = os.getenv('HOME')+'/aiffel/AIFFEL_quest_rs/Exploration/Ex04/images/juun.png'  
# 저장 이미지 위치 지정하기
save_path = os.getenv('HOME') + '/aiffel/AIFFEL_quest_rs/Exploration/Ex04/images/juun_blur.png'

changeBackground(img_path, save_path=save_path)
```

### 동물 배경 블러

```python
# 동물 이미지 위치 불러오기
img_path = os.getenv('HOME')+'/aiffel/AIFFEL_quest_rs/Exploration/Ex04/images/cat.png'  
# 저장 이미지 위치 지정하기
save_path = os.getenv('HOME') + '/aiffel/AIFFEL_quest_rs/Exploration/Ex04/images/cat_blur.png'

changeBackground(img_path,label_name='animal', save_path=save_path)
```

## 문제에 대한 해결방안

LMS에 제시된 해결방안 중 상당수는 3D 카메라, 적외선 센서, 듀얼 카메라 등 촬영 당시의 하드웨어 장비를 활용하여 피사계 심도를 구현하는 방식에 초점을 맞추고 있다. 이러한 접근은 실제 카메라 성능을 개선하여 정확한 심도 정보를 확보하는 데 유효할 수 있지만, 이번 프로젝트의 목적과는 본질적으로 방향이 다르다.

이번 과제는 이미 촬영된 정적인 이미지를 대상으로, 마치 피사계 심도가 적용된 것처럼 후처리를 통해 시각적인 깊이감을 부여하거나 배경을 대체하는 것이 목표였다. 이처럼 후처리에 초점을 둔 문제 상황에서, 촬영 당시의 장비를 전제로 한 해결방안은 적용 가능성이 낮다.

LMS에서 제시된 해결방안 중 하나였던 Google의 Struct2Depth와 같은 단일 이미지 기반의 심도 추정 기술은 예외적으로 주목할 만하다. 이 기술은 기존의 사진만으로도 장면 내의 거리 차이를 추정해낼 수 있기 때문에, 정적인 이미지에서도 피사계 심도에 근접한 시각 효과를 구현할 수 있는 소프트웨어적 대안으로 적합하다고 판단된다.

결론적으로, LMS에 포함된 여러 방안 중 다수는 이번 프로젝트 맥락에서는 적절하지 않았지만, 소프트웨어 기반의 심도 추정 기법은 오히려 문제의 본질에 부합하는 효과적인 해결책이 될 수 있다. 이에 따라 다음과 같은 두 가지 후처리 중심 해결 전략을 제안한다:

1. 경계 자연화: Gaussian Blur 기반 마스크 처리
기존에 사용한 cv2.blur는 경계가 급격하게 끊겨 인물과 배경 사이의 이질감을 유발했다. 이를 개선하기 위해 cv2.GaussianBlur를 이용해 마스크를 부드럽게 처리하면, 경계가 점진적으로 전이되는 자연스러운 보케 효과를 구현할 수 있다.

2. 단일 이미지 기반 심도 추정 기법 활용
Struct2Depth, MiDaS와 같은 기술을 활용하면, 단일 이미지에서 각 객체의 상대적 거리를 추정할 수 있다. 이를 segmentation 결과와 결합해 실제 심도 정보를 반영한 계층적 블러 처리 또는 배경 대체가 가능하다. 이 방식은 단순한 semantic segmentation보다 더 입체감 있고 현실적인 후처리 결과를 기대할 수 있다.
